```python
import random
from imdb import IMDb

genres = ['Documentary', 'History', 'Family',
          'Fantasy', 'Mystery', 'Romance', 
          'Sci-Fi','Thriller', 'Western']

# Pick one random genre
# random_genre = random.choice(genres)
# print('üéØ Target Genre:', random_genre)

ia = IMDb()
timestry = 30

for attempt in range(timestry):
    random_id = str(random.randint(1, 9999999))
    try:
        movie = ia.get_movie(random_id)
        full_info = ia.get_movie(movie.movieID)

        movie_genres = full_info.get('genres', [])

        if movie_genres and any(g in genres for g in movie_genres):
            print(f"‚úÖ Match found on attempt {attempt+1}")
            print(f"üé¨ {full_info['title']} ({full_info.get('year')})")
            print(f"Genres: {', '.join(movie_genres)}")
            break
    except Exception:
        continue
else:
    print("‚ùå No movie found matching the target genre after several attempts.")


full_info = ia.get_movie(movie.movieID)
print(f"Title: {full_info['title']}")
print(f"Year: {full_info.get('year')}")
print(f"Genres: {', '.join(full_info.get('genres', []))}")
print(f"Plot: {full_info.get('plot')[0]}")
print(f"Directors: {', '.join([d['name'] for d in full_info.get('directors', [])])}")
print(f"Cover URL: {full_info.get('cover url')}")
```
